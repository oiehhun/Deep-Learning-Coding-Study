# Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('penguins.csv',index_col=0)
df.head()
```

```python
# Shape of data
df.shape
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측치 제거
df = df.dropna().reset_index(drop=True)
```

```python
# 결측치 확인
df.isnull().sum()
```

연도 변수는 독립 변수에서 제외

```python
df = df.iloc[:,:-1]
df.head()
```

```python
df.shape
```

```python
# Convert categorical variable into dummy

island_dummies = pd.get_dummies(df.island)
sex_dummies = pd.get_dummies(df.sex)
```

```python
island_dummies.head()
```

## Concatenate dataframe

```python
df = pd.concat([df,island_dummies,sex_dummies],axis=1)
df.head()
```

```python
df = df.drop(['island','sex'],axis=1)
df.head()
```

```python
df.shape
```

## Labels

```python
df.species.value_counts()
```

```python
# 문자열 형태를 정수형으로 변환
def make_int(s):
  if s == 'Adelie':
    return 0
  elif s == 'Gentoo':
    return 1
  else:
    return 2
```

```python
df['species'] = df.species.apply(make_int)
```

```python
df.head()
```

나중에 one-hot encoding을 위해서는 정수형으로 바꿔주어야 함
