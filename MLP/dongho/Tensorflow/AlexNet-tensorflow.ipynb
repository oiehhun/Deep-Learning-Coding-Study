# Module

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Dataset

```python
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
# Label 종류
CLASS_NAMES= ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
print('Shape of X train : ',X_train.shape)
print('Shape of y train : ',y_train.shape)
print('Shape of X test : ',X_test.shape)
print('Shape of y test : ',y_test.shape)
```

Cifar10는 색상 사진이기 때문에 3차원 channel을 가지게 됩니다

```python
# 시각화

plt.figure(figsize=(20,20))
for i, (image,label) in enumerate(zip(X_train[:5],y_train[:5])):
  ax = plt.subplot(5,5,i+1)
  plt.imshow(image)
  plt.title(CLASS_NAMES[label[0]])
  plt.axis('off')
```

AlexNet은 (227,227) 입력 크기를 받기 때문에 크기를 조절해줍니다

```python
X_train = [tf.image.resize(image,(227,227)) for image in X_train]
X_test = [tf.image.resize(image,(227,227)) for image in X_test]
```

```python
# 정규화

X_train = X_train / 255
X_test = X_test / 255
```
